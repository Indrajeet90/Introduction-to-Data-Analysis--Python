# Exploring with Visuals
Use `clean_08.csv` and `clean_18.csv`

```python
# load datasets
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
import plotly.plotly as py

df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
```

```python
df_08['greenhouse_gas_score'].plot(kind='hist', title='Greenhouse gas score in 2008')
```

```python
df_18['greenhouse_gas_score'].plot(kind='hist',color='seagreen', title='Greenhouse gas score in 2018')
```

```python
df_08['cmb_mpg'].plot(kind='hist', title='Combined mpg in 2008')
```

```python
df_18['cmb_mpg'].plot(kind='hist', color='seagreen',title='Combined mpg in 2018')
```

```python
df_08.plot(y='displ',x='cmb_mpg',kind='scatter', color='darkviolet',title='Correlation between displacement and combined mpg')
```

```python
df_08.plot(y='greenhouse_gas_score',x='cmb_mpg',kind='scatter', color='darkturquoise',title='Correlation between greenhouse gas score and combined mpg')
```
